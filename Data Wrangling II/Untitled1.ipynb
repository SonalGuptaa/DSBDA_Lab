```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('academic_performance.csv')
```

```python
df.dtypes
```

```python
col_nums=["math score","reading score","writing score"]
```

```python
for col in col_nums:
    # Create boxplot to visualize outliers
    sns.boxplot(x=df[col])
#     plt.title(col)
    plt.show()
    
    # Calculate interquartile range (IQR)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Replace outliers with median value
    df[col] = np.where(df[col] < Q1 - 1.5*IQR, df[col].mean(),df[col])
    df[col] = np.where(df[col] > Q3 + 1.5*IQR, df[col].mean(),df[col])
```

```python
sns.boxplot(x=df['-'])
plt.show()
```
